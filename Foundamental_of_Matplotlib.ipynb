# Introduction To Matplotlib
Matplotlib is a plot graph library in python, iwdly used for visualization of data. 

We'll go over a few basic plot's for you to understand how data points are ploted.

In the example below, we have two array's or axis's. One for Year and one for population. Each element inside of the array is known as a data point.



````python
import matplotlib.pyplot as plt
````

````python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.scatter(year,pop)  # This only creates the data points on the graph
plt.plot(year,pop)     # This creates a line through these data points
plt.show()
````

# Plotting
We'll explore the different way's of ploting graphs using Matplotlib.

One of few parameter's of plot() are:

```
# The Third Parameter is the DataPoint marker, so showcase that point
plt.plot(array1, array2, 'o') 

```

We are also able to change the line style, the line color and even the size of the marker.


```
plt.plot(ypoints, color = '#fffff'
plt.plot(ypoints, ls = 'dotted'
plt.plot(ypoints, linewidth = '20.5')
```

We can clear a plot by. This clears the previous plot, before ploting the next on the screen
```
plt.show()
plt.clf()
```

# Visualization By Histogram
This type of visual help's us to explore expand the meaning of our dataset by getting an idea about the distribution of the values.

We use hist function to plot a histogram in matplotlib
```
# bin mean's in how many parts the data should be divided in the histogram
plt.hist(<List of Values>,<bins>)
```

We start to understand that the importance of bins in a histogram is immense. 

If the bins are too low, it doesn't show a realistic pattern. If it's too many, then the pattern becomes saturated and hard to detect.
